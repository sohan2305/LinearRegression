```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
knn_data = pd.read_csv(r"C:\Users\Ram Choudhary\OneDrive\Desktop\Data Science\14-K-Nearest-Neighbors\KNN_Project_Data", index_col=0)
```

```python
knn_data.head()
```

```python
sns.pairplot(knn_data,hue='TARGET CLASS', palette='coolwarm')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
sc.fit(knn_data.drop('TARGET CLASS', axis =1))
```

```python
scTransform = sc.transform(knn_data.drop('TARGET CLASS', axis=1))
```

```python
df = pd.DataFrame(scTransform, columns=knn_data.columns[:-1])
```

```python
df.head()
```

```python
# Train Test SPlit
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(scTransform, knn_data['TARGET CLASS'],test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier(n_neighbors=1)
```

```python
kn.fit(x_train,y_train)
```

```python
predicts = kn.predict(x_test)
predicts
```

```python
from sklearn.metrics import classification_report, confusion_matrix 
```

```python
print(classification_report(predicts,y_test))
```

```python
error_rate=[]
for i in range(1,40):
    kn= KNeighborsClassifier(n_neighbors=i)
    kn.fit(x_train, y_train)
    pre = kn.predict(x_test)
    error_rate.append(np.mean(pre!= y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40), error_rate, color='blue', linestyle='dashed',marker='*',
        markerfacecolor='red', markersize=10)
plt.title('Error Rate v/s K value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
kn = KNeighborsClassifier(n_neighbors=1)
kn.fit(x_train,y_train)
prd = kn.predict(x_test)
print('WITH K=1')
print('\n')
print(confusion_matrix(y_test,prd))
print('\n')
print(classification_report(y_test,prd))
```
